# **Direct sampling and Gibbs sampler**

```python
#joblib
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from tqdm import tqdm
```

**Direct sampling**

- Let´s say we have the following regression model:
    \begin{align}
    y_{i} = \beta x_{i} + u_{i},{\ \ u}_{{i\ }}\sim\ N\ (0,\ \sigma^{2})
    \end{align}
- with $\beta\sim\ N\left( \beta_{0},\ \tau^{2} \right)$ prior distribution. how can we get posterior mean and equal-tail-probability credible set for $|\beta|$?

- We know that follows a normal distrubution with the following form: 
\begin{align}
\beta|Y,X\ \sim\ N\ (\frac{\frac{1}{\sigma^{2}}\ \sum_{i = 1}^{N}{y_{i}x_{i} + \frac{1}{\tau^{2}}\ \beta_{0}}}{\frac{1}{\sigma^{2}}\ \sum_{i = 1}^{N}{x_{i}^{2} + \frac{1}{\tau^{2}}}} ,\frac{1}{\frac{1}{\sigma^{2}}\ \sum_{i = 1}^{N}{x_{i}^{2} + \frac{1}{\tau^{2}}}})
\end{align}
-Lets asume $\tau=100$, $\sigma^{2}$=1, $\beta_{0}$=0, $\beta$=-1 and N=200

- The first thing we need to do is simulate data of $y_{i}$ and $x_{i}$. To ensure your results are replicated define a random state seed

```python
rng = np.random.RandomState(2088888)
```

- Next, use numpy to simulate $x_{i}$ and $u_{i}$

```python
X=rng.normal(0, 1, size=(200, 1))
u=rng.normal(0, 1, size=(200, 1))
```

- To simulate $y_{i}$ use $x_{i}$, $u_{i}$ and $\beta$=-1

```python
B_real=-1
Y=((X)*B_real)+u
```

$\sum_{i = 1}^{N}$

- Let´s define:
Z= $\sum_{i = 1}^{N}$ $y_{i}$ $x_{i}$ and $\sum_{i = 1}^{N}$ \ \{x_{i}^{2}

```python
Z=Y.T.dot(X)
H=X.T.dot(X)
```

- Replacing $\tau$, $\sigma^{2}$ and $\beta_{0}$ we get:
\begin{align}
\beta|Y,X\ \sim\ N\ (\frac{\sum_{i = 1}^{N}{y_{i}x_{i} }}{\sum_{i = 1}^{N}{x_{i}^{2} + \frac{1}{10000}}} ,\frac{1}{\sum_{i = 1}^{N}{x_{i}^{2} + \frac{1}{10000}}})
\end{align}
-With this expresion we can simulate $\beta$

```python
mu=(Z)/(H+(1/10000))
sigma=(1)/(H+(1/10000))
```

```python
B=rng.normal(mu, sigma, size=(1000, 1))
```

```python
B.mean()
```

```python
abs(B).mean()
```

- As we can see, the estimated values of $|\beta|$ and $\beta$ are realy close to the expected values

```python
fig, ax = plt.subplots()
kwargs = dict(histtype=u'step', alpha=0.3,  bins=40)
ax.hist(B, **kwargs)


##fig.savefig("histogramas.png")
```

- How to get the 90\% equal-tail-probability interval?

```python
np.quantile(B, 0.05)
```

```python
np.quantile(B, 0.95)
```

```python
NS=[100,1000,10000,100000,1000000,10000000]
error=[]
X=rng.normal(0, 1, size=(100, 1))
u=rng.normal(0, 1, size=(100, 1))
B_real=-1
Y=((X)*B_real)+u
Z=Y.T.dot(X)
H=X.T.dot(X)
mu=(Z)/(H+(1/10000))
sigma=(1)/(H+(1/10000))
for i in tqdm(NS):
    B=rng.normal(mu, sigma, size=(int(i), 1))
    er=(np.var(B))
    error.append(e)
    
    
```

```python
data=pd.DataFrame({'N':NS,'Error':error})
```

# Gibbs Sampler

```python
rng = np.random.RandomState(2083)
```

```python
X=rng.normal(0, 1, size=(100, 1))
u=rng.normal(0, 1, size=(100, 1))
B_real=-1
Y=((X)*B_real)+u
```

```python
fig, ax = plt.subplots()
ax.scatter(X,Y, c="g", alpha=0.5, marker=r'$\clubsuit$',
            label="Luck") 
ax.set(xlabel='X', ylabel='Y',
       )
ax.grid()
```

```python
b0=0
a=np.array([10])
b=np.array([20])
e=(Y-X*b0)**2
err=sum(e)
```

```python
def betan(sigma,y,x):
    s_x2 =X.T.dot(X)
    s_xy =Y.T.dot(X)
    w=(1/sigma)*s_x2 + 1
    w =1/w
    m=((1/sigma)*sum(s_xy))*w
    m= np.array([m])
    w = np.array([w])
    beta=rng.normal(m,w)
    return(beta)
```

```python
betas=[]
sigmas=[]
iss=[]
errores=[]
a_hat=a+(100/2)
for i in tqdm(range(1,10000)):
    iss.append(i)
    b_hat=(b+((1/2)*(err)))**-1
    sigma=(np.random.gamma(a_hat,b_hat))**-1
    sigmas.append(sigma)
    beta=betan(sigma,Y,X)
    betas.append(beta)
    e=(Y-X*beta)**2
    err=sum(e[0])
    errores.append(err)
```

```python
sigmas
```

```python
betass=[]
for i in tqdm(betas):
    betass.append(float(i))
```

```python
sigmass=[]
for i in tqdm(sigmas):
    sigmass.append(float(i))
```

```python
fig, ax = plt.subplots()
kwargs = dict(histtype=u'step', alpha=0.3,  bins=40)
ax.hist(sigmass, **kwargs)
```

```python
fig, ax = plt.subplots()
kwargs = dict(histtype=u'step', alpha=0.3,  bins=40)
ax.hist(betass, **kwargs)
```

```python
fig, ax = plt.subplots()
ax.scatter(iss,sigmass, c="g", alpha=0.5, marker=r'$\clubsuit$',
            label="Luck") 
ax.set(xlabel='i', ylabel='Sigma',
       )
ax.grid()
```

```python
fig, ax = plt.subplots()
ax.scatter(iss,betass, c="g", alpha=0.5, marker=r'$\clubsuit$',
            label="Luck") 
ax.set(xlabel='i', ylabel='Beta')
ax.grid()
```
